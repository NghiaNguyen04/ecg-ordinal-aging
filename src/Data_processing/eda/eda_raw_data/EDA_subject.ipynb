#### 2. Kiểm tra độ cân bằng các nhóm tuổi

```python
import matplotlib.pyplot as plt
# Vẽ biểu đồ cột

age_group_counts = subject_raw['Age_group'].value_counts()

fig, ax = plt.subplots() # Tạo figure và axes object để dễ dàng tùy chỉnh
bars = ax.bar(age_group_counts.index, age_group_counts)

# Thêm nhãn số (value_counts) lên trên mỗi cột
for bar in bars:
    # Lấy chiều cao của cột (chính là value_counts)
    height = bar.get_height()
    # Thêm văn bản (nhãn) tại vị trí (x, y) = (trung tâm của cột, chiều cao)
    ax.text(bar.get_x() + bar.get_width() / 2., # Vị trí x: giữa cột
            height,                             # Vị trí y: ngay trên cột
            '%d' % int(height),                 # Nội dung văn bản (số nguyên)
            ha='center',                        # Căn chỉnh ngang: căn giữa
            va='bottom'                         # Căn chỉnh dọc: ở dưới (để văn bản nằm trên cột)
            )

# Đặt tiêu đề và nhãn cho trục (nên có)
ax.set_title('Số lượng theo Nhóm tuổi')
ax.set_xlabel('Nhóm tuổi')
ax.set_ylabel('Số lượng')

# Hiển thị biểu đồ
plt.show()
```
